```python
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
train=pd.read_csv("train.csv")
```

## Working on Train dataset

```python
train.head()
```

```python
train.shape
```

```python
train.isnull().sum().sum()
```

```python
train.dtypes
```

```python
X_train= train.drop(columns=["y", "ID"])
```

```python
categorical_col = [x for x in X_train.columns if train[x].dtype == "object"]
```

```python
# Fetching Categorical Columns
categorical_col
```

```python
# Fetching Numerical Columns
numerical_col = [x for x in X_train.columns if train[x].dtype != "object"]
```

#### Scaling on Train dataset

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
```

```python
X_train_num_scale = pd.DataFrame(scale.fit_transform(X_train[numerical_col]), columns=numerical_col)
```

```python
X_train_num_scale
```

#### Removing those features having zero variance

```python
for col in numerical_col:
    if X_train_num_scale[col].var()<=0:
        X_train_num_scale.drop(columns=col, inplace=True)
```

```python
X_train_num_scale.head(2)
```

### One hot Encoding on Train Dataset

```python
X_train_OHE = pd.get_dummies(train, columns=categorical_col)
```

```python
X_train_OHE.head(2)
```

```python
X_train_final=pd.concat([X_train_OHE, X_train_num_scale], axis=1)
```

```python
X_train_final.head(2)
```

```python
X_train_final.to_csv("Training Data Final.csv")
```

## Working on Test Data

```python
pd.set_option("display.max_columns", None)
test=pd.read_csv("test.csv")
```

```python
test.head()
```

```python
test.shape
```

```python
test.isnull().sum().sum()
```

```python
test.dtypes
```

```python
X_test=test.drop(columns=["ID"])
```

```python
X_test.head(1)
```

```python
# Fetching Numerical & Categorical columns of test data
test_cat_col = [x for x in X_test.columns if X_test[x].dtype == "object"]
test_num_col = [x for x in X_test.columns if X_test[x].dtype != "object"]
```

```python
X_test_num_scale = pd.DataFrame(scale.fit_transform(X_test[test_num_col]), columns=test_num_col)
```

```python
X_test_num_scale.shape
```

#### Removing features having zero variance (test data)

```python
for col in test_num_col:
    if X_test_num_scale[col].var()<=0:
        X_test_num_scale.drop(columns=col, inplace=True)
```

```python
X_test_num_scale.shape
```

```python
X_test_OHE = pd.get_dummies(test, columns=test_cat_col)
```

```python
X_test1 = pd.concat([X_test_OHE, X_test_num_scale], axis=1)
```

```python
X_test1.head(1)
```

###  PCA

```python
X_train_final.shape, X_test1.shape
```

### As we can see the test columns having more columns than the test data

```python
train_columns = set(X_train_final.columns)
test_columns = set(X_test1.columns)

# Find the common columns
common_columns = list(train_columns.intersection(test_columns))

```

```python
X_train_final=X_train_final[common_columns]
X_test1=X_test1[common_columns]
```

```python
X_test1.shape, X_train_final.shape
```

```python
# Fill Missing values with 0
X_test1.fillna(0, inplace=True)
```

```python
X_test1.isnull().sum().sum()
```

## PCA

```python
from sklearn.decomposition import PCA
PCA=PCA(n_components=95)
```

```python
train_PCA = pd.DataFrame(PCA.fit_transform(X_train_final))
```

```python
train_PCA.head(2)
```

```python
PCA.explained_variance_ratio_.sum()
```

```python
test_PCA = pd.DataFrame(PCA.fit_transform(X_test1))
test_PCA.head(2)
```

# XGBoost

```python
y_train = train["y"]
```

```python
from xgboost import XGBRegressor
```

```python
xgb=XGBRegressor()
xgb.fit(train_PCA, y_train)
```

```python
yhat=xgb.predict(test_PCA)
```
