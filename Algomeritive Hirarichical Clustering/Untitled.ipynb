```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel("old data- application form csv.xlsx", sheet_name="Sheet1")
```

```python
clrs = [ 'grey' if (x < max(df.value_counts())) else 'red' for x in df.value_counts()]

plt.figure(figsize=(16,6), )
plt.title("Subjects Counts for Class 11th (old)")
plt.xlabel("Counts")
plt.ylabel("Subjects")
df.value_counts().sort_values(ascending=False).plot(kind="barh" ,color=clrs)
```

```python
df1=pd.read_excel("New Data - Class 11th.xlsx", sheet_name="Sheet3")
```

```python
clrs = [ 'blue' if (x < max(df1.value_counts())) else 'red' for x in df1.value_counts()]

plt.figure(figsize=(30,15), )
plt.title("Subjects Counts for Class 11th (new)")
plt.xlabel("Counts")
plt.ylabel("Subjects")
df1.value_counts().sort_values(ascending=False).plot(kind="barh" ,color=clrs)
```
