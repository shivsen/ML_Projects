### K MEANS Clustering

```python
X=df[["satisfaction_level", "last_evaluation", "left"]]
```

```python
X.shape
```

```python
wcss=[]
for i in range(1,8):
    kmeans=KMeans(n_clusters=i, init="k-means++", random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,8), wcss, c="r", marker="o")
```

```python
kmeans=KMeans(n_clusters=3, init="k-means++", random_state=42)
X["label"]=kmeans.fit_predict(X)
```

```python
print("average silhoutte Score", silhouette_score(X, kmeans.labels_))
```

```python
a=pd.get_dummies(df["sales"], drop_first=True)
```

```python
df=pd.concat([df, a], axis=1)
```

```python
df.drop(columns=["sales"], inplace=True)
```

```python
df.salary.replace({"low" : 0, "medium" : 1, "high" : 2}, inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns="left")
y=df["left"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=123, train_size=0.8)
```

```python
y_train.value_counts()
```

```python
from imblearn.combine import SMOTETomek
```

```python
smote=SMOTETomek()
```

```python
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

```python
y_train_res.value_counts()
```

# Logistic Regression

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
grid = {"penalty" : ['l2', 'elasticnet'],
    "C" : [0.1, 1, 10]
}
```

```python
cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
Grid_search = GridSearchCV(lr, param_grid=grid, cv=cv, scoring="accuracy")
```

```python
Grid_search.fit(X_train_res, y_train_res)
```

```python
Grid_search.best_score_
```

```python
Grid_search.best_params_
```

```python
yhat=Grid_search.predict(X_test)
```

```python
print("classification Report of Logistic Regression : \n ", classification_report(yhat, y_test))
print("confusion matrix : \n", confusion_matrix(yhat, y_test))
print("accuracy \n", accuracy_score(yhat, y_test))
```

```python
lr_prob = Grid_search.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
lr_fpr, lr_tpr, lr_threshold = roc_curve(y_test, lr_prob)
```

```python
lr_roc_score = roc_auc_score(y_test, lr_prob)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(lr_fpr, lr_tpr, lw=2, label='ROC curve (AUC = %0.2f)' % lr_roc_score)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")

# Add threshold values to plot
num_thresholds = len(lr_threshold)
step_size = int(num_thresholds/10) + 1
for i in range(0, num_thresholds, step_size):
    plt.text(lr_fpr[i], lr_tpr[i], '%0.2f' % lr_threshold[i], fontsize=12,
             horizontalalignment='center', verticalalignment='center')
    
# Shade area under curve
plt.fill_between(lr_fpr, lr_tpr, 0, alpha=0.2, color='b')

plt.show()
```
